```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('cat.jpeg') 
plt.imshow(image[:,:,::-1])
plt.title("RGB")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.title("BGR")
plt.show()
```

```python
medium = np.median(image)
medium
```

```python
low = int (max(0,0.7*medium))
high = int (max(255,1/3*medium))
low
```

```python
high
```

```python
blur_img = cv2.blur(image, ksize=(3,3))
plt.imshow(blur_img[:,:,::-1])
plt.title("BLUR")
plt.show()
```

```python
canny = cv2.Canny(blur_img, threshold1= low, threshold2=high )
plt.imshow(canny,cmap='gray')
plt.title("Canny Edge Img")
plt.show()
```
